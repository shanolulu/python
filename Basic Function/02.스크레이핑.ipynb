# 정규 표현식(RE)
https://developer.mozilla.org/ko/docs/Web/JavaScript/Guide/%EC%A0%95%EA%B7%9C%EC%8B%9D

```
import re
p = re.compile('a|b')
m = p.findall('abcdefg')
print(m)
```

```
# 위에 두 가지를 한 번에
# (patten, value)
m1 = re.findall('^Life', 'Life is too short') # ^시작하는 단어
m2 = re.findall('^is', 'Life is too short')
print('m1=', m1)
print('m2=', m2)

m3 = re.findall('short$', 'Life is too short') # 끝나는 단어$
m4 = re.findall('short$', 'Life is too short. So What?')
print('m3=', m3)
print('m4=', m4)
```

```
p = re.compile(r'(\w+)\s+(\d+[-]\d+[-]\d+)')
m = p.search('lee 010-1234-1234')
print(m.group())
print(m.group(1))
print(m.group(2))
```

```
import urllib.request
import re

url = 'https://finance.naver.com/item/main.nhn?code=017670'
html = urllib.request.urlopen(url)
html_contents = str(html.read().decode('ms949'))
print(html_contents)
```

```
# 첫 번째 html 패턴
# <dl class='blind'>
# <dd>xxxxxxxxxxxx</dd>
# </dl>xxxxxxxxxxxxxx</dd>
stock_results = re.findall('(\<dl class=\"blind\"\>)([\s\S]+?)(\</dl\>)', html_contents) # [\s]+ 단어가 들어온다. ? 여러개
print(stock_result)

stock_info = stock_results[0]
stock_index = stock_info[1]
print(stock_info)
print(stock_index)
```

```
# 주식 정보만 추출
index_list = re.findall("(\<dd\>)([\s\S]+?)(\<\/dd\>)", str(stock_index))
```
