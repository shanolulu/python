# MapReduce
Map()

reduce() - 현재는 파이썬 기본 함수에서 빠짐

```
# map
ex = [1,2,3,4,5]
f = lambda x : x ** 2
result = map(f, ex)
print('Map result1=', result)
print('Map result2=', list(result))
```

```
# list comprehension

multiples = [n ** 2 for n in ex]
print('Map result3=', list(multiples))
```

```
# filtering
result4 = map(lambda x: x ** 2 if x % 2 == 0 else x, ex)
print('Map result4=', list(result4))
```

```
result5 = [x ** 2 if x % 2 == 0 else x for x in ex]
print('Map result5=', result5)
```

```
# reduce 데이터를 통합(2개의 값을 넣으면 하나의 출력값이 나옴)시키며, 결과적으로 모든 값을 통합시킴
from functools import reduce
result6 = reduce(lambda x, y: x+y, ex)
print('Reduce result6=', result6)
```

```
x=0
for y in ex:
    x +=y 
print('Reduce result7=', x)
```

```
# reduce를 사용해 최대값 구하기
mydata = [-2, -3, -5, 16, 3]

# 1) function 활용하기
def mymax(data):
    value = 0
    for e in data:
        if value < e:
            value = e
    return value
print(mymax(mydata))

# 2) reduce 활용하기
reduce_result = reduce(lambda x, y: x if x > y else y, mydata)
print(reduce_result)
```
