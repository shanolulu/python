```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('한글 테스트')
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x, y, 'r--') # red -선
plt.subplot(2,2,2)
plt.plot(y, x, 'b*-') # red *-선
plt.subplot(2,1,2)
plt.plot(x, x, 'yx') # red x
```

```python
# Matplotlib Object Oriented Model
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, weight (range 0 to 1)
# right, top은 1-left, 1-bottom
axes.plot(x,y,'b')
axes.set_xlabel('x좌표')
axes.set_ylabel('y좌표')
axes.set_title('제목')
```

```python
fig = plt.figure() # empty canvas

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # insert axes

# Large Figure Axes1
axes1.plot(x, y, 'b')
axes1.set_xlabel('axes1_x좌표')
axes1.set_ylabel('axes1_y좌표')
axes1.set_title('axes1_제목')
# Insert(small) Figure Axes2
axes2.plot(x, y, 'y')
axes2.set_xlabel('axes2_x좌표')
axes2.set_ylabel('axes2_y좌표')
axes2.set_title('axes2_제목')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
for ax in axes:
    ax.plot(x,y,'b')
    ax.set_xlabel('x좌표')
    ax.set_ylabel('y좌표')
    ax.set_title('제목')
fig
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1]) # 0.1 0.1 0.8 0.8
ax.plot(x, x ** 2, label='x**2')
ax.plot(x, x ** 3, label='x**3')
ax.legend() # legend는 라벨이 있어야 출력가능하다
```

```python
fig, ax = plt.subplots()

ax.plot(x, x**2, 'b.-')
ax.plot(x, x**3, 'g--')
```

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color='blue')
ax.plot(x, x+2, color='#FF8C22')
ax.plot(x, x+3, color='#7A113C')
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green")
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title('Defalut axes ranges')

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title('Defalut axes ranges')

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title('Defalut axes ranges')
```
