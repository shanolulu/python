```
matrix = np.arange(1,7) # vector 형태로 만들어짐(출력시 리스트형태이지만 value들을 ,로 나누지 않고 있음)
print(matrix)

# reshape(차수를 변경)
matrix = np.arange(1,7).reshape(2,3)
print(matrix)
print(matrix.sum()) # value들의 합
print(matrix.sum(axis=0)) # row에 대해서
print(matrix.sum(axis=1)) # colum에 대해서
```

```
# 3차원 데이터 = tensor
# list = [[[1,2,3], [1,2,3]], [[1,2,3], [1,2,3]]]
tensor = np.arange(1, 19).reshape(3,2,3)
print(tensor)
print('SUM = ', tensor.sum(axis=None))
print('SUM = ', tensor.sum(axis=0))
print('SUM = ', tensor.sum(axis=1))
print('SUM = ', tensor.sum(axis=2))
```

```
# mean
matrix = np.arange(1, 7).reshape(2,3)
print(matrix.mean())
print(matrix.mean(axis=0))
print(matrix.mean(axis=1))
```

```
# 수직(vstack) 세로 axis = 0

matrix1 = np.array([1,2,3]) #1x3
matrix2 = np.array([4,5,6]) #1x3
print(matrix1)
print(matrix2)
print(np.vstack((matrix1, matrix2))) # 2 x 3
```

```
# 수평(hstack) 가로 axis = 1
matrix1 = np.array([1,2,3]).reshape(3,1) #1x3 -> 3x1
matrix2 = np.array([4,5,6]).reshape(3,1) #1x3 -> 3x1
print(matrix1)
print(matrix2)
print(np.hstack((matrix1, matrix2))) # 3 x 2
```

```
matrix1 = np.array([1,2,3]) #1x3
matrix2 = np.array([4,5,6]) #1x3
print(np.concatenate((matrix1, matrix2), axis=0))

matrix3 = np.array([1,2,3]).reshape(3,1) #1x3 -> 3x1
matrix4 = np.array([4,5,6]).reshape(3,1) #1x3 -> 3x1

print(np.concatenate((matrix3, matrix4), axis=1))
```
