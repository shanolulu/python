# 열과 행 추가

```
import pandas as pd
product_list = [
                {'name': 'Mouse', 'price':100, 'compant':'A'},
                {'name': 'SSD', 'price':200, 'compant':'B'},
                {'name': 'CPU', 'price':300, 'compant':'C'}
]

df = pd.DataFrame(product_list, columns=['name', 'price', 'compant'])
df
```

```
df['expires'] = 5
df
```

```
series = pd.Series([10, 2, 2], index = [0,1,2])
df['count'] = series
df
```

```
df['total'] = df['price'] * df['count']
df
```

```
df.info()
```

```
df.describe()
```

```
discount = []
for row in df['total']:
    if row >= 1000:
        discount.append('10%')
    elif row >= 500:
        discount.append('5%')
    else:
        discount.append('0%')

df['discount'] = discount
df
```

```
def is_wireless(row):
    if row == 'Mouse':
        return True
    else:
        return False
    
df['wireless'] = df.name.apply(is_wireless)
df
```

```
created_date = ['2004-08-09', '2009-08-31', '2020-09-01']
df['created_date'] = created_date
df
```

```
# Quiz
# 년도만 마지막 컬럼에 추가 (columnname = year, ex) 2004

year = []
for i in df['created_date']:
    year.append( i.split('-')[0] )

df['year'] = year
df
```

```
df['created_date']
```

```
# 데이터 값 변경
df['expires'] = 7
df
```

```
# Numpy와 Pandas 같이..
import numpy as np
# 2010보다 연도가 클 경우 0.95를 곱하고 아닌경우 그대로 출력한다.
df['price'] = np.where(df['year']>'2010', df['price'] * 0.95, df['price'])
df
```
