```python
# Quiz ~ ex.py
import pandas as pd

df = pd.read_csv('04_01_data.csv')
df.head()
```

```python
# Q1 서울, 대전, 부산 컬럼으로 데이터프레임 생성
df2 = df[['서울', '대전', '부산']]

df2.head()
```

```python
# Q2 32, 23, 8, 10 행만을 출력
df3 = df2.loc[[32, 23, 8, 10]]
df3
```

```python
# Q3 인덱스 변경 -> 교통범죄, 사기, 절도범죄, 폭행
df3.index = ['교통범죄', '사기', '절도범죄', '폭행']
df3
```

```python
# Q4 각 범죄별 전체 지역 합계
df3['합계'] = df3['서울'] + df3['대전'] + df3['부산']
df3
```

```python
# Q5 서울과 부산에서 1만건 이상의 범죄 조회
df[(df['서울'] >= 10000) & (df['부산'] >= 10000) ]
```

```python
# Q6 행 <--> 열 바꾸기
df2 = df.T
df2.drop(['범죄대분류', '범죄중분류', '계'], inplace=True)
df2.columns = df['범죄중분류']
df2
```

```python
# Q7 범죄명의 데이터를 이용해서 조회
df2[df2['절도범죄'].astype(int) > 10000]
```
