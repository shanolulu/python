```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
##1. csv 파일 불러오기 (convenient_store.csv)
df = pd.read_csv('convenient_store.csv')
df
```

```python
##2. 전체 컬럼 정보, null 값 유무 확인
# print(df.isnull().sum())
# print(df.isnull())
df.info()
```

```python
##3. 개수, 평균, 편차, 최소, 최대값 확인
df.describe()
```

```python
##4. 지역에 대한 통계, 개수, 유니크한 정보, 제일 빈도가 높은 지역
df['area'].describe()
```

```python
##5. 시간 당 급여가 6500원 이상인 지역의 편의점 정보 출력 (상위 10개만)
df[df['hourly_wage'] >= 6500]['company'].head(10)
```

```python
##6. 시간 당 급여가 높은 순서로 정렬 (sort_value() 함수 사용, 상위 10개만 출력)
df[df['hourly_wage'] >= 6500]['company'].sort_values().head(10)
```

```python
##7. 영등포구에서 시간 당 급여가 6000원 이상인 편의점 검색
df[(df['hourly_wage'] > 6000) & (df['area1'] == '영등포구') ]
```

```python
##8. CU 편이점만 출력 (상위 10개만)
df[df['company'].apply(lambda x: x.split(' ')[0]) == 'CU'].head(10)
```

```python
##9. 지역 컬럼(location)을 추가한 다음, in Seoul 이라는 값 저장, 상위 5개 출력
df['location'] = 'in Seoul'
df.head(5)
```

```python
##10. 6000원 이상 컬럼 추가(more_than_6000) -> True, False 값 저장 (상위 20개 출력)
df['more_than_6000'] = (df['hourly_wage'] >= 6000)
df.head(20)
```

```python
##11. more_than_6000 컬럼에서 True인 데이터들의 평균, 개수, 편차 등의 정보 출력

# sum(df['more_than_6000'] == True)
df[df['more_than_6000'] == True].describe()
```

```python
##12. more_than_6000 이름의 함수를 생성하고, 6000원이상인 경우 A group, 아니면 B group을 반환하는 함수 생성

def more_than_6000(pay):
    if pay >= 6000:
        return 'A group'
    else:
        return 'B group'
    
df['hourly_wage'].apply(more_than_6000)
```

```python
##13. more_than_6000_f 컬럼 생성하고 more_than_6000 함수의 결과를 저장
def more_than_6000(pay):
    if pay >= 6000:
        return 'A group'
    else:
        return 'B group'
    
df['more_than_6000_f'] = df['hourly_wage'].apply(more_than_6000)
```

```python
##14. 지금까지의 결과 상위 10개를 출력
df.head(10)
```

```python
##15-1. more_than_6000가 True인 데이터의 지역과 시간당 급여를 가진 새로운 데이터프레임 생성(data2)
##15-2. data2 데이터를 시간당 급여 순으로 정렬 (높은순)
df2 = df[df['more_than_6000'] == True][['area', 'hourly_wage']]
df2

df2['hourly_wage'] = sorted(df2['hourly_wage'], reverse = True)
df2
```

```python
##16. data2를 darta2.csv 파일로 저장
df2.to_csv('data2.csv',index=None, encoding='cp949') # ms949
df3 = pd.read_csv('data2.csv',encoding='cp949')
df3.head()
```

```python
##17. 시간당 급여를 histogram 으로 표시 (matplotlib hist() 사용)

bins = 100 # 구간을 두고 본다는 것을 의미

x = df['hourly_wage']
plt.hist(x, bins=bins)

plt.show()
```
